```python
def validate_op(expr):
    ops = ["<=", ">=", "==", "!="]
    return any(op in expr for op in ops)
print(validate_op("balance >= 5000"))  # True
def check_speed(expr):
    # This checks for ++ or -- presence
    return "++" in expr or "--" in expr
print(check_speed("speed++"))   # True
print(check_speed("speed+++"))  # True (invalid in strict grammar)
```
